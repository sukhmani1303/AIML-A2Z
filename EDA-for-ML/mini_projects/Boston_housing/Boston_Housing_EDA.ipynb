```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"/content/drive/MyDrive/BostonHousing.csv")
df.head()
```

```python
from sklearn.decomposition import PCA
```

```python
pc = PCA(n_components = 10)
```

```python
pc 
```

```python
model = pc.fit(df)
```

```python
model.explained_variance_ratio_
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.distplot(model.explained_variance_ratio_)
```

```python
categorical_var=df.drop(['crim', 'zn', 'indus',  'nox', 'rm', 'age', 'dis',  'tax',
       'ptratio', 'b', 'lstat'],axis=1)
continuous_var=df.drop([ 'chas', 'rad'],axis=1)
```

```python
for i in categorical_var:
    if i != 'medv':
        sns.catplot(x=i,data=categorical_var,kind='count')
        sns.catplot(x=i,y='medv',data=categorical_var,kind='box')
```

```python

for i in continuous_var:
    if i != 'medv':
        
        sns.distplot(continuous_var[i])
        plt.show()
```

```python
for i in continuous_var:
    if i != 'medv':
        sns.scatterplot(x=i,y='medv',data=continuous_var)
        
        plt.show()
```
